```python
import pandas as pd

file = './data/KBO_2019_player_gamestats.csv'
data = pd.read_csv(file, encoding='cp949')
data.head()
```

```python
data['상대'].unique() #@가 붙어있으면 원정경기, 없으면 홈경기
```

```python
opp_list = []
home_away_list = []

for opp in data['상대']:
    if "@" in opp:
        home_away = '원정'
        opp = opp.replace('@', '')
    else:
        home_away = '홈'
    opp_list.append(opp)
    home_away_list.append(home_away)
    
data['상대팀'] = opp_list
data['홈어웨이'] = home_away_list
data.head()
```

```python
team = data.pivot_table(index = ['팀','이름','생일','상대팀'],
                       values = ['타수','안타','홈런','루타','타점','볼넷','사구','희비'],
                       aggfunc = 'sum')
team.head()
```

```python
#상대팀별 타수가 0보다 큰 경우
team = team[ team['타수']>0 ]
team.head()
```

```python
team = team.reset_index()
team.head()
```

```python
# 타율 : 타격에 성공해서 진루하는 비율 = 안타/타수
# 출루율 : 살아서 진루하는 비율 = (안타+볼넷+몸데맞는볼)/(타수+볼넷+몸에맞는볼+희생플라이)
# 장타율 : 타율에 진루한 베이스 가중치 추가 = 루타/타수
# OPS : 출루율 + 장타율

def cal_hit(df):
    df['타율'] = df['안타']/df['타수']
    df['출루율'] = (df['안타']+df['볼넷']+df['사구'])/(df['타수']+df['볼넷']+df['사구']+df['희비'])
    df['장타율'] = df['루타']/df['타수']
    df['OPS'] = df['출루율'] + df['장타율']
    
    return df
```

```python
stats_opp = cal_hit(team)
stats_opp.head()
```

```python
#특정 팀(두산)을 상대로 타수가 10보다 많은 선수 중 출루율 상위 10명
team = '두산'
stats_opp[ (stats_opp['팀']=='두산') & (stats_opp['타수']>10) ].sort_values(by='출루율', ascending=False).head(10)
```

```python
#전체 팀 대상으로 타수가 20보다 큰 선수의 팀별 출루율 상위 5인 타자
ss_df = pd.DataFrame() #빈 데이터프레임 생성

for team in stats_opp['상대팀'].unique():
    s = stats_opp[ (stats_opp['상대팀']==team) & (stats_opp['타수']>20)]
    ss = s.sort_values(by = '출루율', ascending = False).head(5)
    
    ss_df = ss_df.append(ss)
    
ss_df
```

```python
#특정팀 상대 출루율 5위 이내 들어있는 타자 리스트 (중복제거)
ss_df['이름'].unique()
```

```python
#특정팀 상태 출루율 5위 이내 상위타자들을 대상으로 팀별 출루율 피벗테이블
# df['칼럼명'].isin(리스트) : 칼럼 값이 리스트 안에 속하면 True, 아니면 False
'''
condition = []
for name in stats_opp['이름']:
    if name in ss_df['이름'].unique():
        condition.append(True)
    else:
        condition.append(False)
        
stats_opp[condition]
'''
#위의 방법을 판다스의 isin함수 이용
condition = stats_opp['이름'].isin(ss_df['이름'].unique())
top = stats_opp[condition]

top_pivot = top.pivot_table(index = ['팀','이름'], values = '출루율', columns = '상대팀')
top_pivot
```
