# 4. 데이터 표 병합하기

```python
fpath = './data/exam.xlsx'
A = pd.read_excel(fpath, index_col='번호')
A
```

```python
fpath2 = './data/exam_extra.xlsx'
B = pd.read_excel(fpath2, index_col='번호')
B
```

```python
# pd.merge(a, b, how = , left_on= , right_index= )
```

```python
# A,B 테이블을 A테이블에 있는 키 값을 기준으로(how='left')
# A테이블은 인덱스를 키 값으로, B테이블은 인덱스를 키 값으로 병합
pd.merge(A,B,how='left',left_index=True, right_index=True)
```

```python
# B테이블에 있는 키 값을 기준으로(how='right')
# A테이블은 인덱스를 키 값으로, B테이블은 인덱스를 키 값으로 병합
pd.merge(A,B,how='right',left_index=True, right_index=True)
```

```python
# A,B테이블 양쪽에 모두 존재ㅏ는 키 값을 기준으로(how='inner')
pd.merge(A,B,how='inner',left_index=True, right_index=True)
```

```python
# A,B테이블 중 한번이라도 존재하는 모든 키 값을 기준으로(how='outer')
pd.merge(A,B,how='outer',left_index=True, right_index=True)
```

# 5. 저장하기 

```python
# df.to_excel('파일경로+파일명.xlsx', index=False)
```

```python
total = pd.merge(A,B,how='left',left_index=True, right_index=True)
total
```

```python
fpath = './data/exam_total.xlsx'
total.to_excel(fpath)
```

```python
# 판다스의 데이터프레임(df)는 항상 인덱스 값을 가짐
# 인덱스 저장하고 싶지 않으면 index = False 옵션 사용
fpath = './data/exam_total_withoutindex.xlsx'
total.to_excel(fpath, index=False)
    
```

```python
# 가장 대표적인 이름은?
raw.head()
```

```python
raw['YearOfBirth']
```

```python
#세대기준으로 그룹 만들기
year_class_list = []
for year in raw['YearOfBirth']:
    if year <= 1930:
        year_class = '1930년이전'
    elif year <= 1960:
        year_class = '1960년이전'
    elif year <= 1990:
        year_class = '1990년이전'
    else:
        year_class = '2020년이전'
        
    year_class_list.append(year_class)
    
```

```python
raw['year_class'] = year_class_list
raw.head()
```

```python
name_period = raw.pivot_table( index=['Name','Sex'], values='Number', aggfunc='sum', columns='year_class')
name_period = name_period.fillna(0) #Nan을 0으로 저장
name_period = name_period.astype(int) #실수를 정수로 바꿈
name_period.head()
```

```python
#세대 전체 합계
#방법1
name_period['1930년이전'] + name_period['1960년이전'] + name_period['1990년이전'] + name_period['2020년이전']
#방법2
name_period.sum() #행별 합계 (axis=0)
name_period['sum'] = name_period.sum(axis=1) #컬럼별 합계

name_period.head()
```

```python
#세대별 등록비율
for col in name_period.columns:
    col_new = col + '비율'
    name_period[col_new] = name_period[col]/name_period['sum']
name_period.head()    
```

```python
#이름사용수합계, 2020년이전비율, 1990년이전비율 기준으로 내림차순
name_period = name_period.sort_values(by=['sum','2020년이전비율','1990년이전비율'], ascending = False)
name_period.head()
```

```python
name_period.index
```

```python
name_period = name_period.reset_index() #인덱스를 새로 지정
name_period.head()
```
