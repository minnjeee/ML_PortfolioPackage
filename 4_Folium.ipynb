## 지도 생성하기 

```python
import folium
```

## 정보추가하기  

```python
# m = folium.Map(location=[위도,경도], zoom_start=확대정도)
#서울역([37.5536067, 126.9674308])기준으로 지도생성
m = folium.Map(location=[37.5536067, 126.9674308], zoom_start=10)
m
```

```python
# 마커 추가하기
# folium.Marker([위도,경도]).add_to(m)
folium.Marker([37.5536067, 126.9674308]).add_to(m)
m
```

```python
folium.Marker([37.5536067, 126.9674308], tooltip='mouseover').add_to(m) #마우스를 위치하면 나타나는 정보 표시
m
```

```python
folium.Marker([37.5536067, 126.9674308], popup='click').add_to(m) #마우스로 클릭하면 보이는 정보 표시
m
```

```python
# 원 추가하기
# CircleMarker([위도,경도], radius=원크기조절).add_to(m)
folium.CircleMarker([37.5536067, 126.9674308], radius=20).add_to(m)
m
```

```python
# 미니맵 추가하기
from folium.plugins import MiniMap
```

```python
m = folium.Map(location=[37.5536067, 126.9674308], zoom_start=12)
minimap = MiniMap()
minimap.add_to(m)
m
```

## ex. 서울 대피소 현황지도 만들기 

```python

#https://data.seoul.go.kr/dataList/OA-21063/S/1/datasetView.do

import pandas as pd

file = './data/서울시지진대피소.csv'
f = pd.read_csv(file, encoding = 'cp949') #utf-8 // MS-office --> cp949
f.head()
```

```python
f.info()
```

```python
#인덱스번호가 0인 경우, 위도/경도/수용시설명 데이터 출력

i = 0 #인덱스번호

lat = f.loc[i, '위도']
long = f.loc[i, '경도']
name = f.loc[i, '수용시설명']

print(lat, long, name)
```

```python
#모든인덱스 번호에 대해 출력

for i in range(len(f)):
    lat = f.loc[i, '위도']
    long = f.loc[i, '경도']
    name = f.loc[i, '수용시설명']

    print(lat, long, name)
```

```python
#지도 생성하기(중심위치: 서울역 [37.5536067, 126.9674308])
m = folium.Map(location = [37.5536067, 126.9674308], zoom_start = 12)

#대피소 마커추가
for i in range(len(f)):
    lat = f.loc[i, '위도']
    long = f.loc[i, '경도']
    name = f.loc[i, '수용시설명']

    folium.Marker([lat,long], tooltip=name).add_to(m)
m
```

```python
#지도저장
m.save('./data/Sheltermap.html')
```

```python
## 마커 많을 때, ClusterMarker를 이용해 근처에 있는 마커들끼리 그룹으로 표현 
from folium.plugins import MarkerCluster
```

```python
#지도 생성하기(중심위치: 서울역 [37.5536067, 126.9674308])
m = folium.Map(location = [37.5536067, 126.9674308], zoom_start = 12)

marker_cluster = MarkerCluster().add_to(m) #마커를 그룹화
minimap.add_to(m) #미니맵 추가

#대피소 마커추가
for i in range(len(f)):
    lat = f.loc[i, '위도']
    long = f.loc[i, '경도']
    name = f.loc[i, '수용시설명']

    folium.Marker([lat,long], tooltip=name).add_to(marker_cluster)
    
m
```

```python
#지도저장
m.save('./data/Sheltermap2.html')
```
