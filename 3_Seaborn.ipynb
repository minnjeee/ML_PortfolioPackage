# SEABORN

```python
import seaborn as sns
```

```python
# seaborn에 있는 예제로 활용할 수 있는 데이터셋 불러옴 (tips 데이터셋 활용)
# 불러온 데이터셋은 판다스의 데이터프레임형식 -> 동일한 명령어 사용가능
t = sns.load_dataset('tips')
```

```python
t.head()
```

```python
t.info()
```

## 2개의 수치형 데이터의 그패프

```python
# sns.그래프종류(data=df, x='x축칼럼', y='y축칼럼', hue='색상구별기준칼럼')
# sns.relplot(data=df, x=, y=, hue=, kind='그래프종류') #kind옵션 :scatter'(기본값), 'line'
sns.relplot(data=t, x='tip', y='total_bill') #산점도 
```

```python
sns.relplot(data=t, x='tip', y='total_bill', kind='line') #선그래프
```

```python
sns.relplot(data=t, x='tip', y='total_bill', hue='sex') #성별기준으로 색상구분
```

```python
# sns.jointplot(data=df, x=, y=, kind='그래프형태')
#kind옵션: 'scatter'(point, 기본값), 'reg'(point+regression), 'kde'(누적분포차트 like 지도)
sns.jointplot(data=t, x='tip', y='total_bill')
```

```python
sns.jointplot(data=t, x='tip', y='total_bill', kind='kde')
```

```python
sns.jointplot(data=t, x='tip', y='total_bill', kind='reg')
```

```python
sns.jointplot(data=t, x='tip', y='total_bill', kind='hex')
```

## 모든수치형데이터에서 두개를 선택하여 그래프

```python
# sns.pairplot(data=df, hue=)
sns.pairplot(t)
```

```python
sns.pairplot(data=t, hue='sex')
```

## 수치형과 카테고리형 

```python
# sns.boxplot(data=df, x=, y=, hue=)
sns.boxplot(data=t, x='day', y='tip')
```

```python
sns.boxplot(data=t, x='day', y='tip', hue='smoker')
```

```python
# sns.swarmplot(data=df, x=, y=) #데이터개수와 함께 분포 확인가능
sns.swarmplot(data=t, x='day', y='tip')
```

```python
sns.swarmplot(data=t, x='day', y='tip', hue='smoker')
```

```python
sns.swarmplot(data=t, x='day', y='tip', hue='smoker', dodge=True)
```

```python
sns.boxplot(data=t, x='day', y='tip', hue='smoker')
sns.swarmplot(data=t, x='day', y='tip', hue='smoker', dodge=True)
```

```python
# 사이즈/성별 별 tip의 분포
sns.boxplot(data=t, x='size', y='tip', hue='sex')
```

```python
# barplot(data=t,x=,y=,hue=)
sns.barplot(data=t, x='size', y='tip')
```

```python
sns.barplot(data=t, x='size', y='tip', hue='sex')
```

## 수치형과 2개의 카테고리형 

```python
# heatmap(data=df, annot=True, fmt='.1f', cmap='RdBu_r')
#cmap옵션 추천색상 : 'Reds', 'Blues', 'vlag', 'Pastel1', 'RdBu_r'
tt =t.pivot_table(index='day',columns='size', values='tip', aggfunc='mean')
tt
```

```python
sns.heatmap(data=tt)
```

```python
sns.heatmap(data=tt, annot=True) #수치표현
```

```python
sns.heatmap(data=tt, annot=True, fmt='.2f') #표현되는 형태 지정(fommating)
```

```python
sns.heatmap(data=tt, annot=True, fmt='.0f') #표현되는 형태 지정(fommating)-정수로표현 
```

```python
sns.heatmap(data=tt, annot=True, fmt='.2f', cmap='Blues') #색상조정
```
