```python
import pandas as pd
import os
```

```python
os.chdir(r'./data') #경로 설정
```

```python
iris = pd.read_csv('IRIS.csv.xls')
iris.head()
```

```python
iris['species'].value_counts()
```

```python
iris['id'] = range(len(iris)) #각각을 구별하기 위해 추가
iris = iris[['id','sepal_length','sepal_width','petal_length','petal_width','species']] #컬럼순서 변경
iris.head()
```

# KNN 실습(분류-범주형예측) 

```python
train = iris.sample(100, replace=False, random_state=2020).reset_index(drop=True)
train
# replace=False : 비복원추출
# random_state : 난수 고정
```

```python
test = iris.loc[ ~iris['id'].isin(train['id']) ].reset_index(drop=True)
test.head()
# ~ : not의 의미, True면 False를, False이면 True를 반환
# x.isin(y) : x가 y에 속하면 True, 속하지 않으면 False를 반환 
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier( n_neighbors=3 ) # k=3인 KNN 객체 생성 
```

```python
# knn.fit( train's input variables, train's target variables )
knn.fit(train[['sepal_length','sepal_width','petal_length','petal_width']], train['species']) #학습시키기
```

```python
# knn.predict(test's input variables)
predictions = knn.predict( test[['sepal_length','sepal_width','petal_length','petal_width']]) #예측하기
predictions
```

```python
test['pred'] = predictions
test.head()
```

```python
(test['species'] == test['pred']).mean() #잘 예측된 비율
```

```python
# 최적의 k 찾기
```

```python
for k in range(1,30):
    knn = KNeighborsClassifier( n_neighbors=k ) # k=3인 KNN 객체 생성 
    knn.fit(train[['sepal_length','sepal_width','petal_length','petal_width']], train['species']) #학습시키기
    predictions = knn.predict( test[['sepal_length','sepal_width','petal_length','petal_width']]) #예측하기
    test['pred'] = predictions
    print((test['species'] == test['pred']).mean()) #잘 예측된 비율
```

```python
#예측비율이 가장 높은 것이 여러개이면, 그 중에서 k가 가장 작은 것을 선택!
# => k=3일 때 최적이다
```

```python
# cross validation 활용하여 최적의 k 찾기
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
for k in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=k)
    print(cross_val_score(knn, iris[['sepal_length','sepal_width','petal_length','petal_width']], iris['species'], cv=3).mean())
```

```python
#=>최적의 k는 4이다
```

# KNN 실습(회귀-연속형예측)

```python
del train['species']
del test['species']
train.head()
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor(n_neighbors=3)
```

```python
knn.fit( train[['sepal_length','sepal_width','petal_length']], train['petal_width'])
```

```python
test['pred'] = knn.predict( test[['sepal_length','sepal_width','petal_length']])
test.head()
```

```python
# mae : 회귀문제에서 모델의 성능을 평가하는 방법
# = (예측치-실제값)의 절대값의 평균

abs(test['petal_width']-test['pred']).mean()
```

```python
# => 0.168만큼 예측치랑 실제값이 떨어져있다 
```

```python
# 최적의 k 찾기
for k in range(1,30):
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit( train[['sepal_length','sepal_width','petal_length']], train['petal_width'])
    test['pred'] = knn.predict( test[['sepal_length','sepal_width','petal_length']])
    print(abs(test['petal_width']-test['pred']).mean())
```

```python
# 가장 작은 것이 최적이다!
# => 최적의 k는 8이다.
```
